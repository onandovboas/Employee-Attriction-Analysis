# Transformação de um Flat CSV em uma modelagem de dados

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Análise de Dados/ibm-hr.csv')

```

```python
# Separei o grande arquivo em 3 arquivos, unidos por contexto de dados
# Testei diferentes métodos para fazer o slicing
demografic_df = df.loc[:,['EmployeeNumber', 'Age', 'Education', 'EducationField', 'Gender', 'MaritalStatus', 'NumCompaniesWorked']]
workInfo_df = df.iloc[:,[9,1,2,3,4,5,12,14,15,18,19,23,24,26,28,29,31,32,33,34]]
workPerception_df = df.iloc[:,[9,10,13,16,25,30]]

newTables = [demografic_df, workInfo_df, workPerception_df]
labelTable = ['demografic_df', 'workInfo_df', 'workPerception_df']

for i in range(len(newTables)):
  newTables[i].to_csv(f'{labelTable[i]}.csv', index=False)

# Outra maneira mais eficaz de fazer seria usando dicionários

createTable = {
    'demografic_df': demografic_df,
    'workInfo_df': workInfo_df,
    'workPerception_df': workPerception_df
}

for name, tabela in createTable.items():
  tabela.to_csv(f'{name}.csv', index=False)
```
