# Transformação de um Flat CSV em uma modelagem de dados

```python
import pandas as pd

# --- FUCTIONS ---
# Fuction to slice a huge dataFrama with their index, it's quicker than write one column for one
def columnList(dataFrame, columnsIndex):
  columnsList = []
  for x in columnsIndex:
    columnsList.append(dataFrame.columns[x])
  return columnsList

# --- VARIABLES ---
# Creating a list with the Columns names
workInfoIndex = [9,1,2,3,4,5,12,14,15,18,19,23,24,26,28,29,31,32,33,34]
workPerceptionIndex = [9,10,13,16,25,30]

# DataFrame
df = pd.read_csv('/content/drive/MyDrive/Análise de Dados/ibm-hr.csv')
```

```python

demografic_df = df.loc[:, ['EmployeeNumber', 'Age', 'Education', 'EducationField', 'Gender', 'MaritalStatus', 'NumCompaniesWorked']]
workInfo_df = df.loc[:, columnList(df,workInfoIndex)]
workPerception_df = df.loc[:, columnList(df, workInfoIndex)]
```

```python
createTable = {
    'demografic_df': demografic_df,
    'workInfo_df': workInfo_df,
    'workPerception_df': workPerception_df
}

for name, table in createTable.items():
  table.to_csv(f'{name}.csv', index=False)
```
